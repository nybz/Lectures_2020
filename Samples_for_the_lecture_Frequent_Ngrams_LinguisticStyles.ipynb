```python
interviews = pd.read_json('C://Users//2918261//Dropbox//Corner Office Interviews/Presentations/Lecture_UiS/interviews.json')
```

```python
text = "".join(interviews.text)
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize, word_tokenize
words = word_tokenize(text)
```

```python
wordsFiltered = [word.lower() for word in words if word.isalpha()]
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
filtered_words = [word for word in wordsFiltered if word not in stopwords.words('english')]
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt 

wc = WordCloud(max_words = 100, margin = 10, background_color = 'purple',
              scale = 3, relative_scaling = 1, width = 500, height =  400, 
              random_state = 1).generate(' '.join(filtered_words))

plt.figure(figsize = (20, 10))
plt.imshow(wc)
plt.axis("off")
plt.show(wc)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt 

wc = WordCloud(max_words = 500, margin = 10, background_color = 'white',
              scale = 3, relative_scaling = 0.5, width = 500, height =  400, 
              random_state = 1).generate(' '.join(filtered_words))

plt.figure(figsize = (20, 10))
plt.imshow(wc)
plt.axis("off")
plt.show(wc)
```
